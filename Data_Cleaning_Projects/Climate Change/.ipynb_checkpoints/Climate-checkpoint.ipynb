#### Summary

Data from World Development Indicators and Climate Change Knowledge Portal on climate systems, exposure to climate impacts, resilience, greenhouse gas emissions, and energy use.

In addition to the data available here and through the Climate Data API, the Climate Change Knowledge Portal has a web interface to a collection of water indicators that may be used to assess the impact of climate change across over 8,000 water basins worldwide. You may use the web interface to download the data for any of these basins.

Here is how to navigate to the water data:

Go to the Climate Change Knowledge Portal home page (http://climateknowledgeportal.worldbank.org/)
Click any region on the map Click a country In the navigation menu
Click "Impacts" and then "Water" Click the map to select a specific water basin
Click "Click here to get access to data and indicators" Please be sure to observe the disclaimers on the website regarding uncertainties and use of the water data.

## Aim

To perform data cleaning on Dataset

#### Importing Libraries

```python
#Python
import numpy as np
import pandas as pd

#Visualization
import matplotlib.pyplot as plt
import seaborn as sns

#Data cleaning
import missingno
```

#### Importing Dataset

The dataset could be found using the following link 

https://query.data.world/s/pb64gfxbsym3sb6qxie3azasbrdyci

```python
climate_df=pd.read_excel("climate_change_download_0.xls")
climate_df.head()
```

```python
climate_df.sample()
```

#### Column Removal

```python
climate_df.columns
```

Here 'Country code' serve same as 'Country name' So we can drop 'Country code'

```python
climate_df1 = climate_df.drop('Country code',axis=1)
climate_df1
```

```python
for i in climate_df1["Series code"].unique():
    print(climate_df1[climate_df1["Series code"]==i]["Series name"].value_counts())
```

```python
climate_df2 = climate_df1.drop('Series name',axis=1)
climate_df2
```

'Series code' is unique for 'Series name' hence we can drop 'Series name'

```python
climate_df3=climate_df2
climate_df3['Identifier']=list(range(0,13512))
climate_df3.head()
```

```python
climate_df3.set_index("Identifier",inplace=True)
climate_df3.head()
```

#### Tidying up fields

```python
climate_df3["Country name"].value_counts()
```

```python
climate_df3["Series code"].value_counts()
```

```python
climate_df3["SCALE"].value_counts()
```

```python
climate_df3["Decimals"].value_counts()
```

#### Missing values and Treatment

```python
climate_df3.isnull().sum()
```

```python
missingno.matrix(climate_df3)
plt.show()
```

It clearly seems that the climate data is missing for some Series code, which won't be helpful. Hence we need to target that Series code

```python
climate_df3[climate_df3[1990].isnull()]['Series code'].value_counts()
```

```python
climate_df4=climate_df3.dropna(subset=climate_df3.iloc[4:-1].columns)
climate_df4.head()
```

```python
climate_df4[climate_df4[1990].isnull()]['Series code'].value_counts()
```

```python
missingno.matrix(climate_df4)
plt.show()
```

```python
climate_df4.isnull().sum()
```

With our data cleaned we will now save it

```python
climate_df4.to_csv("Climate__cleaned.csv")
```
